```python
# @title Remove HTML Tags
import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'',text)
```

```python
df['review'] = df['review'].apply(remove_html_tags)
```

```python
df
```

```python
# @title Remove urls
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)
```

```python
df['review'] = df['review'].astype(str).apply(remove_url)
```

```python
# @title Remove Punctuations
import string,time
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punch(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
df['review'] = df['review'].astype(str).apply(remove_punch)
```

```python
# @title Remove Stopwords
import nltk
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger') # For POS tagging, crucial for accurate lemmatization

def get_wordnet_pos(treebank_tag):
  if treebank_tag.startswith('J'):
      return wordnet.ADJ
  elif treebank_tag.startswith('V'):
      return wordnet.VERB
  elif treebank_tag.startswith('N'):
      return wordnet.NOUN
  elif treebank_tag.startswith('R'):
      return wordnet.ADV
  else:
      return wordnet.NOUN # Default to noun if POS not clearly mapped

def lemmatize_text(text):
  lemmatizer = WordNetLemmatizer()
  # Convert to string to handle potential non-string inputs
  text = str(text)
  tokens = word_tokenize(text)
  lemmatized_words = []
  for word, tag in nltk.pos_tag(tokens):
    pos = get_wordnet_pos(tag)
    lemmatized_words.append(lemmatizer.lemmatize(word, pos=pos))
  return " ".join(lemmatized_words)
```

###Find out the number of words in the entire corpus and also the total number of unique words(vocabulary).

```python
total_words = 0
unique_words = set()

for review in df['review']:
    words = review.split()
    total_words += len(words)
    for word in words:
        unique_words.add(word)

print(f"Total number of words in the corpus: {total_words}")
print(f"Total number of unique words (vocabulary): {len(unique_words)}")
```

###One Hot Encoding


```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer
# You can adjust parameters like max_features to limit the vocabulary size
vectorizer = CountVectorizer()

# Fit and transform the 'review' column
X = vectorizer.fit_transform(df['review'])

# X is a sparse matrix
print("Shape of the Bag-of-Words matrix:", X.shape)

# You can get the vocabulary (list of unique words)
vocabulary = vectorizer.get_feature_names_out()
print("Length of vocabulary:", len(vocabulary))
```

###bag words and find the vocabulary also find the times each word has occured

```python
# Get the vocabulary
vocabulary = vectorizer.get_feature_names_out()

# Sum the occurrences of each word across all reviews
word_counts = X.sum(axis=0)

# Create a dictionary mapping words to their total counts
word_counts_dict = dict(zip(vocabulary, word_counts.tolist()[0]))

# Display the vocabulary and the total counts for the first 20 words as an example
print("Vocabulary and total word counts (first 20 words):")
for word, count in list(word_counts_dict.items())[:20]:
    print(f"{word}: {count}")

# You can also access the count of a specific word like this:
# print(f"\nTotal count of the word 'movie': {word_counts_dict.get('movie', 0)}")
```

###bag of bi-gram and bag of tri-gram and write down your observation about the dimensionality of the vocabulary

```python
from sklearn.feature_extraction.text import CountVectorizer

# Bag of Bigrams
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2))
X_bigrams = bigram_vectorizer.fit_transform(df['review'])

print("Shape of the Bag-of-Bigrams matrix:", X_bigrams.shape)
print("Length of bigram vocabulary:", len(bigram_vectorizer.get_feature_names_out()))

# Bag of Trigrams
trigram_vectorizer = CountVectorizer(ngram_range=(3, 3))
X_trigrams = trigram_vectorizer.fit_transform(df['review'])

print("\nShape of the Bag-of-Trigrams matrix:", X_trigrams.shape)
print("Length of trigram vocabulary:", len(trigram_vectorizer.get_feature_names_out()))
```

###tf-idf and find out the idf scores of words, also find out the vocabulary.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
# You can adjust parameters like max_features to limit the vocabulary size
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the 'review' column
X_tfidf = tfidf_vectorizer.fit_transform(df['review'])

# Get the vocabulary and IDF scores
vocabulary = tfidf_vectorizer.get_feature_names_out()
idf_scores = tfidf_vectorizer.idf_

# Create a dictionary mapping words to their IDF scores
idf_scores_dict = dict(zip(vocabulary, idf_scores))

# Display the vocabulary and the IDF scores for the first 20 words as an example
print("Vocabulary and IDF scores (first 20 words):")
for word, score in list(idf_scores_dict.items())[:20]:
    print(f"{word}: {score:.4f}")

# You can also access the IDF score of a specific word like this:
# print(f"\nIDF score of the word 'movie': {idf_scores_dict.get('movie', 0)}")
```
